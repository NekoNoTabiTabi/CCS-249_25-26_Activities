#### Manual Implementation

```python
## Import Libraries
import math
import re
from collections import defaultdict
```

```python
## Define Dataset
documents = [
    ("Free money now!!!",                       "SPAM"),
    ("Hi mom, how are you?",                    "HAM"),
    ("Lowest price for your meds",              "SPAM"),
    ("Are we still on for dinner?",             "HAM"),
    ("Win a free iPhone today",                 "SPAM"),
    ("Let's catch up tomorrow at the office",   "HAM"),
    ("Meeting at 3 PM tomorrow",                "HAM"),
    ("Get 50% off, limited time!",              "SPAM"),
    ("Team meeting in the office",              "HAM"),
    ("Click here for prizes!",                  "SPAM"),
    ("Can you send the report?",                "HAM")
]

testText = [
    "Limited offer, click here!",
    "Meeting at 2 PM with the manager."
]

## Tokenization
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())
```

```python
## Bag of Words

# Count Frequency
spam_frequency = defaultdict(int)
ham_frequency = defaultdict(int)
vocabulary = set()

spam_total = 0
ham_total = 0
spam_docs_count = 0
ham_docs_count = 0

for doc, label in documents:
    tokens = tokenize(doc)
    if label == "SPAM":
        spam_docs_count += 1
        for token in tokens:
            spam_frequency[token] += 1
            vocabulary.add(token)
            spam_total += 1
            # if new count, checks token and adds vocabulary.
    else:
        ham_docs_count += 1
        for token in tokens:
            ham_frequency[token] += 1
            vocabulary.add(token)
            ham_total += 1

v_size = len(vocabulary)

print("BAG OF WORDS")
print(f"Vocabulary size: {v_size}")
print(f"Total words in SPAM: {spam_total}")
print(f"Total words in HAM: {ham_total}")
```

```python
## Priors
# P(class) = (number of documents in class) / (total documents)

total_docs = len(documents)
p_spam = spam_docs_count / total_docs
p_ham = ham_docs_count / total_docs

print("PRIORS")
print(f"P(SPAM) = {spam_docs_count}/{total_docs} = {p_spam:.4f}")
print(f"P(HAM) = {ham_docs_count}/{total_docs} = {p_ham:.4f}\n")
```

```python
## Likelihood
# P(word|class) = (count of word in class + 1) / (total words in class + vocabulary size)
print("LIKELIHOOD")
for word in vocabulary:
    p_word_given_spam = (spam_frequency[word] + 1) / (spam_total + v_size)
    p_word_given_ham = (ham_frequency[word] + 1) / (ham_total + v_size)
    print(f"P({word}|SPAM) = ({spam_frequency[word]} + 1) / ({spam_total} + {v_size}) = {p_word_given_spam:.4f}")
    print(f"P({word}|HAM) = ({ham_frequency[word]} + 1) / ({ham_total} + {v_size}) = {p_word_given_ham:.4f}")
```

```python
## Determining the Given Class (testText)
print("\nCLASSIFICATION")
for text in testText:
    # Tokenize
    tokens = tokenize(text)
    print(f"\nTest sentence: '{text}'")
    print(f"Tokens: {tokens}")

    # Initialize Prior Probabilities
    prob_spam = p_spam
    prob_ham = p_ham
    print(f"Initial P(SPAM) = {prob_spam:.4f}, P(HAM) = {prob_ham:.4f}")

    # Multiply Likelihoods for Each Token
    for token in tokens:
        p_token_spam = (spam_frequency[token] + 1) / (spam_total + v_size)
        p_token_ham = (ham_frequency[token] + 1) / (ham_total + v_size)
        prob_spam *= p_token_spam
        prob_ham *= p_token_ham
        print(f"Token '{token}': P({token}|SPAM) = {p_token_spam:.4f}, P({token}|HAM) = {p_token_ham:.4f}")

    # Determine Predicted Class
    predicted_class = "SPAM" if prob_spam > prob_ham else "HAM"
    print(f"Total P(SPAM|doc) ∝ {prob_spam:.10f}, Total P(HAM|doc) ∝ {prob_ham:.10f}")
    print(f"Predicted Class: {predicted_class}")
```

#### Using Scikit-Learn

```python
## Import Library
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
## Define Dataset
docText = [
    "Free money now!!!",
    "Hi mom, how are you?",
    "Lowest price for your meds",
    "Are we still on for dinner?",
    "Win a free iPhone today",
    "Let&#39;s catch up tomorrow at the office",
    "Meeting at 3 PM tomorrow",
    "Get 50% off, limited time!",
    "Team meeting in the office",
    "Click here for prizes!",
    "Can you send the report?"
]

classLabel = [
    "SPAM",
    "HAM",
    "SPAM",
    "HAM",
    "SPAM",
    "HAM",
    "HAM",
    "SPAM",
    "HAM",
    "SPAM",
    "HAM"
]

testText = [
    "Limited offer, click here!",
    "Meeting at 2 PM with the manager."
]
```

```python
## Vectorize text (Bag of Words)
vectorizer = CountVectorizer(token_pattern=r"(?u)\b\w+\b")  # remove English stopwords
X_train = vectorizer.fit_transform(docText)

print(vectorizer.get_feature_names_out())
print()
```

```python
## Train Naive Bayes classifier
clf = MultinomialNB(alpha=1.0)

clf.fit(X_train, classLabel)

X_test = vectorizer.transform(testText)
predictions = clf.predict(X_test)
probabilities = clf.predict_proba(X_test)
classes = clf.classes_
```

```python
print("Prediction Result using SciKit-Learn")
for i, sentence in enumerate(testText):
    print(f"\nTest Sentence {i+1}: '{sentence}'")
    print("Probability Distribution:")
    for j, c in enumerate(classes):
        print(f"P({c}): {probabilities[i][j]:.8f}")
    print(f"Predicted Class: {predictions[i]}")
```
