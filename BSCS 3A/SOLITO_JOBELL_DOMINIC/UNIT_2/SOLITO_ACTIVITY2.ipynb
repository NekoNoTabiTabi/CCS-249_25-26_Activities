# **1. Updating ELIZA**

```python
import re

def reflect(fragment):
    """Reflects user input to make responses more natural."""
    reflections = {
        "am": "are",
        "was": "were",
        "i": "you",
        "i'd": "you would",
        "i've": "you have",
        "i'll": "you will",
        "my": "your",
        "are": "am",
        "you've": "I have",
        "you'll": "I will",
        "your": "my",
        "yours": "mine",
        "you": "me",
        "me": "you"
    }
    words = fragment.lower().split()
    return ' '.join([reflections.get(word, word) for word in words])

def eliza_response(user_input):
    """Generates ELIZA-style responses based on input."""
    patterns = [
        (r"I need (.*)", "Why do you need {0}?"),
        (r"Why don't you (.*)", "Do you really think I don't {0}?"),
        (r"I feel (.*)", "Tell me more about feeling {0}."),
        (r"I want to know the reason why I am feeling depressed all the time", "Why do you think you are feeling depressed all the time?"),
        (r"I am feeling stressed", "What do you think is causing you to feel stressed?"),
        (r"My feelings towards my crush are invalidated", "Why do you feel your feelings are being invalidated?"),
        (r"You don't understand me", "I am trying my best to understand. Can you explain it differently?"),
        (r"I can't focus on my studies", "What do you think is distracting you from your studies?"),
    ]
    
    for pattern, response in patterns:
        match = re.match(pattern, user_input, re.IGNORECASE)

        if match:
            print(match.group(1)) 
            return response.format(reflect(match.group(1)))
    
    return "Can you tell me more?"

print("ELIZA: Hello! How can I help you today?")
last_input = ""

while True:
    user_input = input("You: ").strip()
    
    if user_input.lower() in ["quit", "exit"]:
        print("ELIZA: Goodbye!")
        break

    if user_input != "" and user_input == last_input:
        print("ELIZA: Are we echoing now? You literally just said that.")
    else:
        print(f"ELIZA: {eliza_response(user_input)}")
        
    last_input = user_input
```

# **2a. Implementing RegEx on NLP - Upper Case Letter**

```python
import re

text = """Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do.
Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, "and what is the use of a book," thought Alice, "without pictures or conversations?"""

pattern = r"\b[A-Z][a-z]*\b"
uppercase_words = re.findall(pattern, text)

print(uppercase_words)
```

# **2b. Implementing RegEx on NLP - Moby Dick**

```python
import re

pattern = r"\b[Ww]hales?\b"

with open("melville-moby_dick.txt", "r", encoding="utf-8") as file:
    moby_dick_text = file.read()

extracted_words = re.findall(pattern, moby_dick_text)[:10]
for word in extracted_words:
    print(f"{word} -> leviathan")

modified_text = re.sub(pattern, "leviathan", moby_dick_text, count=10)
```

# **2c. Implementing RegEx on NLP - NLTK package webtext**

```python
import nltk
import re

nltk.download('webtext')
from nltk.corpus import webtext
pirates_text = webtext.raw('pirates.txt')

pattern = r"^JACK SPARROW:.*$"

jack_sparrow_lines = re.findall(pattern, pirates_text, re.MULTILINE)

for line in jack_sparrow_lines[:5]:
    print(line)
```
