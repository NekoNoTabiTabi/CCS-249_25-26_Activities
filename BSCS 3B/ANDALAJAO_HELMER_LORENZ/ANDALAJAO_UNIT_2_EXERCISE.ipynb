```python
import re

def reflect(fragment):
    """Reflects user input to make responses more natural."""
    reflections = {
        "am": "are",
        "was": "were",
        "i": "you",
        "i'd": "you would",
        "i've": "you have",
        "i'll": "you will",
        "my": "your",
        "you've": "I have",
        "you'll": "I will",
        "yours": "mine",
        "you": "me",
        "me": "you",
        "to": "for",
    }
    words = fragment.lower().split()
    return ' '.join([reflections.get(word, word) for word in words])

def eliza_response(user_input, previous_inputs):
    """Generates ELIZA-style responses based on input."""
    
    # Check if user is repeating themselves
    if user_input.lower() in previous_inputs:
        return "Again? are you mad?"
    
    patterns = [
        (r"I need (.*)", "Why do you need {0}?"),
        (r"Why don't you (.*)", "Do you really think I don't {0}?"),
        (r"I feel (.*)", "Tell me more about feeling {0}."),
        (r"I want to know the reasons why I am feeling (.*)", "Why are you feeling {0}?"),
        (r"I am feeling (.*)", "What makes you feel {0}?"),
        (r"My feelings towards my (.*)", "Why do you have those feelings about your {0}?"),
        (r"You don't understand (.*)", "I do understand {0}."),
        (r"You do not understand (.*)", "Can you help me understand {0}?"),
        (r"I can't focus on (.*)", "Why can't you focus on {0}?"),
        (r"I cannot focus on (.*)", "Why can't you focus on {0}?"),
    ]
    
    for pattern, response in patterns:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            return response.format(reflect(match.group(1)))
    
    return "Can you tell me more?"

# Track previous inputs
conversation_history = []

print("ELIZA: Hello! How can I help you today?")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["quit", "exit"]:
        print("ELIZA: Goodbye!")
        break
    
    response = eliza_response(user_input, conversation_history)
    print(f"ELIZA: {response}")
    
    # Add to history (only keep last 5 inputs to avoid memory issues)
    conversation_history.append(user_input.lower())
    if len(conversation_history) > 5:
        conversation_history.pop(0)
```

```python
# ELIZA implementation in Python
# Example Generated via ChatGPT
import re
from collections import Counter

with open('Dick.txt', 'r') as file:
	text = file.read()

matches = re.findall(r'[Ww]hale',text)
WordCount = Counter(matches)
print('all whale instances')
print(matches)
print('Number of times the word "whale" was stated')
print(WordCount)
```

```python
pirates_text = webtext.raw('pirates.txt')
lines = pirates_text.split('\n')
jack_sparrow_lines = [line for line in lines if re.match(r'^JACK SPARROW:.*$', line)]

print(f"Total lines spoken by Jack Sparrow: {len(jack_sparrow_lines)}\n")
print("Jack Sparrow's lines:")
for i, line in enumerate(jack_sparrow_lines, 1):
    print(f"{i}. {line}")
```
