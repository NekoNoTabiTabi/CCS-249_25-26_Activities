```python
# Define the dataset as a list of (document, class) tuples
dataset = [
    ("Free money now!!!", "SPAM"),
    ("Hi mom, how are you?", "HAM"),
    ("Lowest price for your meds", "SPAM"),
    ("Are we still on for dinner?", "HAM"),
    ("Win a free iPhone today", "SPAM"),
    ("Let's catch up tomorrow at the office", "HAM"),
    ("Meeting at 3 PM tomorrow", "HAM"),
    ("Get 50% off, limited time!", "SPAM"),
    ("Team meeting in the office", "HAM"),
    ("Click here for prizes!", "SPAM"),
    ("Can you send the report?", "HAM"),
]

# Separate documents by class
spam_docs = [doc for doc, label in dataset if label == "SPAM"]
ham_docs  = [doc for doc, label in dataset if label == "HAM"]

print(f"Total documents : {len(dataset)}")
print(f"SPAM documents  : {len(spam_docs)}")
print(f"HAM documents   : {len(ham_docs)}")
print()

# Display the dataset
for doc, label in dataset:
    print(f"[{label}] {doc}")
```

```python
import re

def tokenize(text):
    """Convert text to lowercase and extract alphabetic words only."""
    return re.findall(r'[a-z]+', text.lower())

# Tokenize all documents
tokenized_spam = [tokenize(doc) for doc in spam_docs]
tokenized_ham  = [tokenize(doc) for doc in ham_docs]

# Display tokenized documents
print("SPAM tokens:")
for tokens in tokenized_spam:
    print(f"  {tokens}")

print("\nHAM tokens:")
for tokens in tokenized_ham:
    print(f"  {tokens}")
```

```python
from collections import Counter

def build_bag_of_words(tokenized_docs):
    """Count the frequency of each word across all documents in a class."""
    bow = Counter()
    for tokens in tokenized_docs:
        bow.update(tokens)
    return bow

# Build Bag of Words for each class
spam_bow = build_bag_of_words(tokenized_spam)
ham_bow  = build_bag_of_words(tokenized_ham)

# Build the full vocabulary (all unique words)
vocabulary = sorted(set(spam_bow.keys()) | set(ham_bow.keys()))

print(f"Vocabulary size: {len(vocabulary)}")
print(f"\nVocabulary: {vocabulary}")
print(f"\nSPAM Bag of Words: {dict(spam_bow)}")
print(f"\nHAM Bag of Words:  {dict(ham_bow)}")
```

```python
def calculate_prior(class_docs, total_docs):
    """Calculate the prior probability of a class."""
    return len(class_docs) / total_docs

total_docs = len(dataset)

# Calculate priors
prior_spam = calculate_prior(spam_docs, total_docs)
prior_ham  = calculate_prior(ham_docs, total_docs)

print(f"P(SPAM) = {len(spam_docs)}/{total_docs} = {prior_spam:.4f}")
print(f"P(HAM)  = {len(ham_docs)}/{total_docs} = {prior_ham:.4f}")
```

```python
# Test sentences to classify later
test_sentences = [
    "Limited offer, click here!",
    "Meeting at 2 PM with the manager."
]

# Check which words from test sentences are in the vocabulary and which are dropped
for i, sentence in enumerate(test_sentences, start=1):
    tokens  = tokenize(sentence)
    in_vocab    = [w for w in tokens if w in vocabulary]
    not_in_vocab = [w for w in tokens if w not in vocabulary]

    print(f"Test {chr(96+i)}: \"{sentence}\"")
    print(f"  Tokens:              {tokens}")
    print(f"  Words in vocabulary: {in_vocab}")
    if not_in_vocab:
        print(f"  Words dropped:       {not_in_vocab} (not in training vocabulary)")
    else:
        print(f"  Words dropped:       None")
    print()
```

```python
def calculate_likelihood(bow, vocabulary, alpha=1):
    """
    Calculate the likelihood P(word | class) for every word in the vocabulary.
    Uses Laplace smoothing with smoothing factor alpha (default = 1).
    """
    total_words = sum(bow.values())
    vocab_size  = len(vocabulary)
    likelihoods = {}

    for word in vocabulary:
        count = bow.get(word, 0)
        likelihoods[word] = (count + alpha) / (total_words + alpha * vocab_size)

    return likelihoods, total_words

# Calculate likelihoods for SPAM and HAM
spam_likelihood, spam_total_words = calculate_likelihood(spam_bow, vocabulary)
ham_likelihood, ham_total_words   = calculate_likelihood(ham_bow, vocabulary)

print(f"Total words in SPAM: {spam_total_words}")
print(f"Total words in HAM:  {ham_total_words}")
print(f"Vocabulary size (|V|): {len(vocabulary)}\n")

# Display likelihoods for each word
print(f"{'Word':<15} {'P(w|SPAM)':<15} {'P(w|HAM)':<15}")
print("-" * 45)
for word in vocabulary:
    print(f"{word:<15} {spam_likelihood[word]:<15.4f} {ham_likelihood[word]:<15.4f}")
```

```python
def format_scientific(value):
    """Format a float as n.nnnn × 10^exp notation."""
    if value == 0:
        return "0"
    exp = int(f"{value:.4e}".split("e")[1])
    coeff = value / (10 ** exp)
    return f"{coeff:.4f} × 10^{exp}"

def classify_with_steps(text, prior_spam, prior_ham, spam_likelihood, ham_likelihood, vocabulary):
    """
    Classify a text and show the multiplication formula.
    """
    tokens = tokenize(text)
    words_used = [w for w in tokens if w in vocabulary]

    # --- SPAM computation ---
    prob_spam = prior_spam
    for word in words_used:
        prob_spam *= spam_likelihood[word]

    # Build the formula string
    spam_values = [f"{prior_spam:.4f}"] + [f"{spam_likelihood[w]:.4f}" for w in words_used]
    spam_labels = ["P(SPAM)"] + [f"P({w}|SPAM)" for w in words_used]
    print("  P(SPAM|doc) = " + " × ".join(spam_labels))
    print("              = " + " × ".join(spam_values))
    print(f"              = {format_scientific(prob_spam)}")

    # --- HAM computation ---
    prob_ham = prior_ham
    for word in words_used:
        prob_ham *= ham_likelihood[word]

    ham_values = [f"{prior_ham:.4f}"] + [f"{ham_likelihood[w]:.4f}" for w in words_used]
    ham_labels = ["P(HAM)"] + [f"P({w}|HAM)" for w in words_used]
    print("\n  P(HAM|doc)  = " + " × ".join(ham_labels))
    print("              = " + " × ".join(ham_values))
    print(f"              = {format_scientific(prob_ham)}")

    predicted = "SPAM" if prob_spam > prob_ham else "HAM"
    return predicted, prob_spam, prob_ham

# Classify each test sentence with step-by-step output
for i, sentence in enumerate(test_sentences, start=1):
    print(f"\nTest {chr(96+i)}: \"{sentence}\"")

    predicted, prob_spam, prob_ham = classify_with_steps(
        sentence, prior_spam, prior_ham,
        spam_likelihood, ham_likelihood, vocabulary
    )

    print(f"\n Predicted class: {predicted}")
    print()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Prepare training data
documents = [doc for doc, label in dataset]
labels    = [label for doc, label in dataset]

print("Training documents:", len(documents))
print("Labels:", labels)
```

```python
# Convert text to a Bag of Words matrix using CountVectorizer
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(documents)

# Display the feature names (vocabulary)
print("Feature names (vocabulary):")
print(vectorizer.get_feature_names_out())
print(f"\nFeature matrix shape: {X_train.shape}")

# Train the Multinomial Naïve Bayes classifier
model = MultinomialNB()
model.fit(X_train, labels)

print("\nModel trained successfully!")
```

```python
# Test sentences
test_sentences = [
    "Limited offer, click here!",
    "Meeting at 2 PM with the manager."
]

# Vectorize the test sentences using the same vectorizer
X_test = vectorizer.transform(test_sentences)

# Predict the class
predictions = model.predict(X_test)

# Get prediction probabilities
probabilities = model.predict_proba(X_test)
class_names   = model.classes_

# Display results
for i, sentence in enumerate(test_sentences):
    print(f"Test {chr(97+i)}: \"{sentence}\"")
    for j, cls in enumerate(class_names):
        print(f"     P({cls}) = {probabilities[i][j]:.4f}")
    print(f" Predicted class: {predictions[i]}")
    print()
```
